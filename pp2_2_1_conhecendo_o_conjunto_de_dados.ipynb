### 7. De acordo com a documentação da base de dados, qual o significado dos atributos do tipo "Soil Type"?

São 40 colunas binárias que representam se o exemplo pertence ou não a um tipo de solo específico.

### 7.1 Este atributo é quantitativo ou qualitativo?

Qualitativo, porque representa categorias de solo e não uma medida numérica como "percentual de argila".

### 7.2 Qual o tipo de codificação utilizada para denotar este atributo?

Como existem diversas categorias de solo, foi aplicada a codificação "One-Hot Encoding", que transforma cada categoria em uma nova coluna, atribuindo o valor 1 ou 0 para indicar, respectivamente, a presença ou ausência daquele tipo de solo.

```python
# Separar apenas os atributos contínuos
continuous_features = [
    'Elevation', 'Aspect', 'Slope',
    'Horizontal_Distance_To_Hydrology',
    'Vertical_Distance_To_Hydrology',
    'Horizontal_Distance_To_Roadways',
    'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
    'Horizontal_Distance_To_Fire_Points'
]
```

```python
# Histograma (distribuição e tendência central)
plt.figure(figsize=(16, 12))
for i, col in enumerate(continuous_features, 1):
    plt.subplot(5, 2, i)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribuição de {col}", fontsize=14)
    plt.xlabel(f'Valores de {col}', fontsize=12)
    plt.ylabel('Frequência', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Boxplot (dispersão e outliers)
plt.figure(figsize=(16, 8))
sns.boxplot(data=df[continuous_features], orient="h")
plt.title("Dispersão e tendência central dos atributos contínuos", fontsize=14)
plt.xlabel('Valores', fontsize=12)
plt.show()
```

```python
estatisticas = df[continuous_features].describe().transpose()

estatisticas.rename(columns={
    'count': 'Quantidade',
    'mean': 'Média',
    'std': 'Desvio Padrão',
    'min': 'Mínimo',
    '25%': '1º Quartil',
    '50%': 'Mediana',
    '75%': '3º Quartil',
    'max': 'Máximo'
}, inplace=True)

estatisticas = estatisticas.round(2) #2 casas decimais

print("Medidas Estatísticas dos Atributos Contínuos:\n")
print(estatisticas)
```

```python
# Colunas correspondentes aos tipos de solo e áreas selvagens
soil_cols = [col for col in df.columns if 'Soil_Type' in col]
wilderness_cols = [col for col in df.columns if 'Wilderness_Area' in col]

# Converter de one-hot para contagem (soma das ocorrências de "1")
soil_counts = df[soil_cols].sum().sort_values(ascending=False)
wilderness_counts = df[wilderness_cols].sum().sort_values(ascending=False)
```

```python
# Frequência dos tipos de solo
plt.figure(figsize=(12, 5))
sns.barplot(x=soil_counts.index, y=soil_counts.values, color='lightcoral')
plt.title('Distribuição dos Tipos de Solo')
plt.xlabel('Tipo de Solo')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
```

```python
# Frequência das áreas selvagens
plt.figure(figsize=(8, 5))
sns.barplot(x=wilderness_counts.index, y=wilderness_counts.values, color='gold')
plt.title('Distribuição das Áreas Selvagens')
plt.xlabel('Área Selvagem')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de barras para cada classe do atributo-alvo
sns.countplot(x='Cover_Type', data=df, color='skyblue')
plt.title('Distribuição do Atributo-Alvo (Cover_Type)')
plt.xlabel('Tipo de Cobertura Florestal')
plt.ylabel('Frequência')
plt.show()
```
